# Recursive Fonksiyon (Öz yinelemeli fonksiyon)
Kendi içerisinde kendine atıfta bulunan (doğrudan veya dolaylı olarak çağıran) fonksiyonlara recursive(rekursif) fonksiyon denir.
Döngüsel yaklaışımın bir alternatifi olarak kullanılabilir.

```python
# Döngüsel yaklaşım: 5! = 5*4*3*2*1
def faktoriyel(n):
    fakt = 1  # çarpım toplamı
    for i in range(n,0,-1):
        fakt = fakt*i # yığmalı çarpma
    return fakt

#ana program
print(faktoriyel(5))
print(faktoriyel(4))
```

```python
# Recursive yaklaşım 5! = 5*4! = 5*4*3! = 5*4*3*2! = 5*4*3*2*1! = 5*4*3*2*1
def faktoriyel(n):
    if n<=1:
        return 1
    else:
        return n*faktoriyel(n-1)

#ana program
print(faktoriyel(5))
print(faktoriyel(4))
```

# Nested Functions(İç içe fonksiyonlar)

** Scope(Faaliyet Alanı) ve LEGB(Local - Enclosing - Global - Built-in) Kuralı:**

    Local (Yerel): İlk olarak, Python yerel alandaki değişkenlere bakar. Bir onksiyon içinde tanımlanan değişkenler yereldir.

    Enclosing (Kapsayan): Eğer aranan değişken yerel alanda bulunamazsa, Python bir üst kapsamda (nested functions veya nested class) değişkeni arar.

    Global (Genel): Eğer değişken yerel veya kapsayan alanda bulunamazsa, Python global alanda değişkeni arar. Fonksiyon dışındaki, ancak aynı dosya veya modül içinde tanımlanan değişkenleri içerir.

    Built-in (Yerleşik): Eğer değişken yukarıdaki hiçbir alanda bulunamazsa, Python yerleşik (built-in) ad alanında değişkeni arar.
    
Bu sıra, bir değişkeni bulmak için Python'un nasıl arama yaptığını tanımlar.

```python
my_var = "outer variable"  # global alan (global değişken)

def func_var():
    my_var = "inner variable" # local alan (yerel değişken)
    print(my_var)

# bir fonksiyonun faaliyeti ancak çağrıldığında başlar
func_var()
print(my_var)  # Acaba global alandaki my_var, yoksa local'deki my_var ?
```

```python
# içiçe fonksiyonlarda return ile içteki fonksiyonu döndürebilirmiyiz?
def outer_func(x):
    # en dıştaki kapsayan fonks.(enclosing func.)
    def inner_func(y):
        # en içteki kapsayan fonks.
        def nested_func(z):
            #en içteki fonksiyon (local)
            result = x+y+z
            return result
        return nested_func(5)   # z:5
    return inner_func(3)        # y:3

# çağıralım
print(outer_func(10))           # x:10
```

```python
# closure
# içiçe fonksiyonlarda return ile içteki fonksiyonu döndürebilirmiyiz?
def outer_func(x):
    # en dıştaki kapsayan fonks.(enclosing func.)
    def inner_func(y):
        # en içteki kapsayan fonks.
        def nested_func(z):
            #en içteki fonksiyon (local)
            result = x+y+z
            return result

        return nested_func

    return inner_func

# çağıralım
closure_olustur = outer_func(10)           # x=10
print(closure_olustur(5)(3))
```

```python
# DA8122 Tuğçe
def my_abs(a):
  z = type(a)
  if z == int or z == float:
    if a < 0:
      return -a
    else:
      return a
  elif z == complex:
    mutlak = pow((pow(a.real,2) + pow(a.imag,2)),0.5)
    if mutlak < 0 :
      return -mutlak
    else:
      return mutlak

my_abs(3+4j)
```

```python
# DA8124: my_pow()
def my_pow(number, exponent):
    """Calculate the power of a number"""
    result=1
    if exponent > 0:
        for i in range(exponent):
            result *= number
    elif exponent < 0:
        number = 1 / number
        exponent = -exponent
        for i in range(exponent):
            result *= number
    else: result=1       # 0 için ama yazmasak da olur.
    return result

print(my_pow(2,3))
```
