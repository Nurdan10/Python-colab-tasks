# Recursive Fonksiyon (z yinelemeli fonksiyon)
Kendi ierisinde kendine atfta bulunan (dorudan veya dolayl olarak aran) fonksiyonlara recursive(rekursif) fonksiyon denir.
Dngsel yaklamn bir alternatifi olarak kullanlabilir.

```python
# Dngsel yaklam: 5! = 5*4*3*2*1
def faktoriyel(n):
    fakt = 1  # arpm toplam
    for i in range(n,0,-1):
        fakt = fakt*i # ymal arpma
    return fakt

#ana program
print(faktoriyel(5))
print(faktoriyel(4))
```

```python
# Recursive yaklam 5! = 5*4! = 5*4*3! = 5*4*3*2! = 5*4*3*2*1! = 5*4*3*2*1
def faktoriyel(n):
    if n<=1:
        return 1
    else:
        return n*faktoriyel(n-1)

#ana program
print(faktoriyel(5))
print(faktoriyel(4))
```

# Nested Functions( ie fonksiyonlar)

** Scope(Faaliyet Alan) ve LEGB(Local - Enclosing - Global - Built-in) Kural:**

    Local (Yerel): lk olarak, Python yerel alandaki deikenlere bakar. Bir onksiyon iinde tanmlanan deikenler yereldir.

    Enclosing (Kapsayan): Eer aranan deiken yerel alanda bulunamazsa, Python bir st kapsamda (nested functions veya nested class) deikeni arar.

    Global (Genel): Eer deiken yerel veya kapsayan alanda bulunamazsa, Python global alanda deikeni arar. Fonksiyon dndaki, ancak ayn dosya veya modl iinde tanmlanan deikenleri ierir.

    Built-in (Yerleik): Eer deiken yukardaki hibir alanda bulunamazsa, Python yerleik (built-in) ad alannda deikeni arar.
    
Bu sra, bir deikeni bulmak iin Python'un nasl arama yaptn tanmlar.

```python
# iie fonksiyonlarda return ile iteki fonksiyonu dndrebilirmiyiz?
def outer_func(x):
    # en dtaki kapsayan fonks.(enclosing func.)
    def inner_func(y):
        # en iteki kapsayan fonks.
        def nested_func(z):
            #en iteki fonksiyon (local)
            result = x+y+z
            return result
        return nested_func(5)   # z:5
    return inner_func(3)        # y:3

# aralm
print(outer_func(10))           # x:10
```

```python
# closure
# iie fonksiyonlarda return ile iteki fonksiyonu dndrebilirmiyiz?
def outer_func(x):
    # en dtaki kapsayan fonks.(enclosing func.)
    def inner_func(y):
        # en iteki kapsayan fonks.
        def nested_func(z):
            #en iteki fonksiyon (local)
            result = x+y+z
            return result

        return nested_func

    return inner_func

# aralm
closure_olustur = outer_func(10)           # x=10
print(closure_olustur(5)(3))
```

```python
# DA8122 Tue
def my_abs(a):
  z = type(a)
  if z == int or z == float:
    if a < 0:
      return -a
    else:
      return a
  elif z == complex:
    mutlak = pow((pow(a.real,2) + pow(a.imag,2)),0.5)
    if mutlak < 0 :
      return -mutlak
    else:
      return mutlak

my_abs(3+4j)
```

```python
# DA8124: my_pow()
def my_pow(number, exponent):
    """Calculate the power of a number"""
    result=1
    if exponent > 0:
        for i in range(exponent):
            result *= number
    elif exponent < 0:
        number = 1 / number
        exponent = -exponent
        for i in range(exponent):
            result *= number
    else: result=1       # 0 iin ama yazmasak da olur.
    return result

print(my_pow(2,3))
```
