```python
# çıktısı nedir?
for num in range(2, 10):
    if num == 5:
        break
    print(num, end=" ")
```

```python
for num in range(2, 10):
    if num == 5:
        continue  # döngü başı yapar
    print(num, end=" ")
```

**Task-2:** Sonsuz döngüler gerçek yaşamda nerelerde kullanılır, araştıralım? Bir kullanım örneğini kodlayınız

```python
# ATM, Trafik sinyalinizasyonu, programlanabilir elektronik kartlar....
# Haydi bir trafik sinyalizasyon programı tasarlayalım...
import time
sinyal_list = ["Yeşil", "Sarı", "Kırmızı"]
while True:  # sonsuz döngü
    sensor = input("Lamba rengi.:")
    if sensor == sinyal_list[0]: # "Yeşil"
        print("GEÇ")
        time.sleep(2)  # 2 saniye bekle
        break
    elif sensor == sinyal_list[1]:  # "Sarı"
        print("BEKLE veya HAZIRLAN")
        time.sleep(1)
    else:
        print("DUR")
        time.sleep(2)
```

 **zip**(*iterables, strict=False) fonksiyonu:

    Iterate over several iterables in parallel, producing tuples with an item from each one.

    Example:

```python
for item in zip([1, 2, 3], ['sugar', 'spice', 'everything nice']):
    print(item)
```

```python
text = ['one', 'two', 'three', 'four', 'five']
numbers = [1, 2, 3, 4, 5]
# Bu iki listeyi birleştirelim...
for x,y in zip(text,numbers):
    print(f"{x} : {y}")
```

# İç içe Döngüler(Nested Loops)

```python
# nested loop örneği: çarpım tablosunu elde edelim
# Kural: içteki döngü bitmeden dıştaki döngü başlamaz
# ..
for i in range(1,6): # dış döngü
    for j in range(1,10): #iç döngü
        print(f"{i}*{j}={i*j}")
    print() # içteki döngü bitince bir boş satır bırak
```

```python
for i in range(1,6): # dış döngü
    for j in range(1,10): #iç döngü
        print(f"{i}*{j}={i*j}", end="\t")
    print() # içteki döngü bitince bir boş satır bırak
```

```python
#Aşağıdaki desenleri (patterns) nasıl ekrana basarız?
# iç içe döngü ile nasıl yazdırırız?
for i in range(1,10):
    for j in range(i):
        print(i,end="")
    print()
```
